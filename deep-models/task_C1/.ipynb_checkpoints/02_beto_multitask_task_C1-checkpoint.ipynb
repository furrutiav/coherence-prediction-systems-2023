```python
from transformers import BertTokenizer, BertModel
import torch
```

```python
model_name = "dccuchile/bert-base-spanish-wwm-cased"
beto_model = BertModel.from_pretrained(model_name)
beto_tokenizer = BertTokenizer.from_pretrained(model_name, do_lower_case=False)
e = beto_model.eval()
```

```python
beto_tokenizer.all_special_tokens, beto_tokenizer.all_special_ids
```

```python
from torch.utils.data import Dataset
class DatasetTaskC1Q(Dataset):
    def __init__(self, df, maxlen):
        self.df = df
        self.tokenizer = beto_tokenizer
        self.maxlen = maxlen

    def __len__(self):
        return len(self.df)

    def __getitem__(self, index):
        sentence1 = str(self.df.loc[index, 'Q'])
        sentence2 = str(self.df.loc[index, 'A'])
        
        sentence1 = " ".join(str(sentence1).replace("\n", " ").split())
        sentence2 = " ".join(str(sentence2).replace("\n", " ").split())

        label = self.df.loc[index, 'label']
        label_Q = self.df.loc[index, 'label_Q']
        
        tokens1 = self.tokenizer.tokenize(sentence1) if len(sentence1)>0 else ["[UNK]"]
        tokens2 = self.tokenizer.tokenize(sentence2) if len(sentence2)>0 else ["[UNK]"]

        if len(tokens1) < self.maxlen:
            tokens1 = tokens1 + ['[PAD]' for _ in range(self.maxlen - len(tokens1))]
        else:
            tokens1 = tokens1[:self.maxlen]

        if len(tokens2) < self.maxlen:
            tokens2 = tokens2 + ['[PAD]' for _ in range(self.maxlen - len(tokens2))]
        else:
            tokens2 = tokens2[:self.maxlen]
          
        tokens = ["[CLS]"]+tokens1+["[SEP]"]+tokens2+["[SEP]"]
        tokens_ids = self.tokenizer.convert_tokens_to_ids(tokens)
        tokens_ids_tensor = torch.tensor(tokens_ids)
        attn_mask = (tokens_ids_tensor != 1).long() # [PAD] => 1

        tokens_Q = ["[CLS]"]+tokens1+["[SEP]"]
        tokens_ids_Q = self.tokenizer.convert_tokens_to_ids(tokens_Q)
        tokens_ids_tensor_Q = torch.tensor(tokens_ids_Q)
        attn_mask_Q = (tokens_ids_tensor_Q != 1).long() # [PAD] => 1

        return tokens_ids_tensor, attn_mask, label, tokens_ids_tensor_Q, attn_mask_Q, label_Q
```

```python
from torch.utils.data import DataLoader

train_set = DatasetTaskC1Q(df = A_train, maxlen = 60)
val_set = DatasetTaskC1Q(df = A_val, maxlen = 60)

train_loader = DataLoader(train_set, batch_size = 64, num_workers = 2, shuffle=False)
val_loader = DataLoader(val_set, batch_size = 64, num_workers = 2, shuffle=False)
```

```python
import torch.nn as nn
class C1QClassifier(nn.Module):
    def __init__(self):
        super(C1QClassifier, self).__init__()
        torch.manual_seed(2022)
        
        self.bert_layer = BertModel.from_pretrained(model_name).cuda()
        self.cls_layer = nn.Linear(768, 2).cuda()
        self.cls_layer_Q = nn.Linear(768, 6).cuda()

    def forward(self, seq, attn_masks, seq_Q, attn_masks_Q):

        cont_reps = self.bert_layer(seq, attention_mask=attn_masks)
        cont_reps_Q = self.bert_layer(seq_Q, attention_mask=attn_masks_Q)

        cls_rep = cont_reps.last_hidden_state[:, 0]
        cls_rep_Q = cont_reps_Q.last_hidden_state[:, 0]

        logits = self.cls_layer(cls_rep)
        logits_Q = self.cls_layer_Q(cls_rep_Q)

        return logits, logits_Q
```

```python
import torch.optim as optim

net = C1QClassifier()

weights = torch.tensor([1., 6.5])
criterion = nn.CrossEntropyLoss(weight=weights, reduction='mean').cuda()

criterion_Q = nn.CrossEntropyLoss().cuda()

opti = optim.Adam(net.parameters(), lr = 2e-5)
```

```python
def get_accuracy_from_logits(logits, labels):
    probs = torch.sigmoid(logits)
    soft_probs = probs.argmax(1)
    acc = (soft_probs.squeeze() == labels).float().mean()
    return acc, soft_probs
    
def evaluate(net, criterion, criterion_Q, dataloader):
    net.eval()
    mean_acc, mean_loss = 0, 0
    mean_acc_Q, mean_loss_Q = 0, 0
    count = 0
    tests, preds = [], []
    tests_Q, preds_Q = [], []
    with torch.no_grad():
        for seq, attn_masks, labels, seq_Q, attn_masks_Q, labels_Q in dataloader:
            seq, attn_masks, labels = seq.cuda(), attn_masks.cuda(), labels.cuda()
            seq_Q, attn_masks_Q, labels_Q = seq_Q.cuda(), attn_masks_Q.cuda(), labels_Q.cuda()

            logits, logits_Q = net(seq, attn_masks, seq_Q, attn_masks_Q)

            mean_loss += criterion(logits, labels).item()
            mean_loss_Q += criterion_Q(logits_Q, labels_Q).item()
            
            acc, soft_probs = get_accuracy_from_logits(logits, labels)
            acc_Q, soft_probs_Q = get_accuracy_from_logits(logits_Q, labels_Q)
            
            mean_acc += acc
            mean_acc_Q += acc_Q

            preds += soft_probs.tolist()
            preds_Q += soft_probs_Q.tolist()
            
            tests += labels.tolist()
            tests_Q += labels_Q.tolist()

            count += 1

    return mean_acc / count, mean_loss / count, tests, preds, mean_acc_Q / count, mean_loss_Q / count, tests_Q, preds_Q 

from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import classification_report
```

```python
def train(net, criterion, criterion_Q, opti, train_loader, val_loader, epochs):
    for ep in range(epochs):
        for it, (seq, attn_masks, labels, seq_Q, attn_masks_Q, labels_Q) in enumerate(train_loader):
            opti.zero_grad()  

            seq, attn_masks, labels = seq.cuda(), attn_masks.cuda(), labels.cuda()
            seq_Q, attn_masks_Q, labels_Q = seq_Q.cuda(), attn_masks_Q.cuda(), labels_Q.cuda()

            logits, logits_Q = net(seq, attn_masks, seq_Q, attn_masks_Q)

            loss = criterion(logits, labels)
            loss_Q = criterion_Q(logits_Q, labels_Q)

            loss.backward()
            loss_Q.backward()

            opti.step()

            if (it + 1) % 100 == 0:
                acc, _ = get_accuracy_from_logits(logits, labels)
                acc_Q, _ = get_accuracy_from_logits(logits_Q, labels_Q)
                print("Iteration {} of epoch {} complete. Loss(C1) : {} Train Accuracy(C1) : {} Loss(Q) : {} Train Accuracy(Q) : {}".format(it+1, ep+1, loss.item(), acc, loss_Q.item(), acc_Q))
        
        val_acc, val_loss, tests, preds, val_acc_Q, val_loss_Q, tests_Q, preds_Q = evaluate(net, criterion, criterion_Q, val_loader)

        print(classification_report(tests, preds))
        print(classification_report(tests_Q, preds_Q))
        print("Epoch {} complete! Validation Accuracy(C1) : {}, Validation Loss(C1) : {} Validation Accuracy(Q) : {}, Validation Loss(Q) : {}".format(ep+1, val_acc, val_loss, val_acc_Q, val_loss_Q))
```

```python
epochs = 6
train(net, criterion, criterion_Q, opti, train_loader, val_loader, epochs)
```

```python
repo_name = "beto_ft_multitask_task_C1"
config_file_url = hf_hub_url("furrutiav/"+repo_name, filename="cls_layer.torch")
value = cached_download(config_file_url)
cls_layer = torch.load(value)

config_file_url = hf_hub_url("furrutiav/"+repo_name, filename="cls_layer_Q.torch")
value = cached_download(config_file_url)
cls_layer_Q = torch.load(value)
```

```python
beto_model = BertModel.from_pretrained("furrutiav/"+repo_name).cuda()
beto_tokenizer = BertTokenizer.from_pretrained("furrutiav/"+repo_name, do_lower_case=False)
e = beto_model.eval()
```

```python
 def preproccesing(Q, A, maxlen=60):
        sentence1 = str(Q)
        sentence2 = str(A)
        
        sentence1 = " ".join(str(sentence1).replace("\n", " ").split())
        sentence2 = " ".join(str(sentence2).replace("\n", " ").split())
        
        tokens1 = beto_tokenizer.tokenize(sentence1) if len(sentence1)>0 else ["[UNK]"]
        tokens2 = beto_tokenizer.tokenize(sentence2) if len(sentence2)>0 else ["[UNK]"]

        if len(tokens1) < maxlen:
            tokens1 = tokens1 + ['[PAD]' for _ in range(maxlen - len(tokens1))]
        else:
            tokens1 = tokens1[:maxlen]

        if len(tokens2) < maxlen:
            tokens2 = tokens2 + ['[PAD]' for _ in range(maxlen - len(tokens2))]
        else:
            tokens2 = tokens2[:maxlen]
          
        tokens = ["[CLS]"]+tokens1+["[SEP]"]+tokens2+["[SEP]"]
        tokens_ids = beto_tokenizer.convert_tokens_to_ids(tokens)
        tokens_ids_tensor = torch.tensor(tokens_ids)
        attn_mask = (tokens_ids_tensor != 1).long() # [PAD] => 1

        return tokens_ids_tensor.cuda(), attn_mask.cuda()
```

```python
 def preproccesingQ(Q, maxlen=60):
        sentence1 = str(Q)
        
        sentence1 = " ".join(str(sentence1).replace("\n", " ").split())
        
        tokens1 = beto_tokenizer.tokenize(sentence1) if len(sentence1)>0 else ["[UNK]"]

        if len(tokens1) < maxlen:
            tokens1 = tokens1 + ['[PAD]' for _ in range(maxlen - len(tokens1))]
        else:
            tokens1 = tokens1[:maxlen]

        tokens = ["[CLS]"]+tokens1+["[SEP]"]
        tokens_ids = beto_tokenizer.convert_tokens_to_ids(tokens)
        tokens_ids_tensor = torch.tensor(tokens_ids)
        attn_mask = (tokens_ids_tensor != 1).long() # [PAD] => 1

        return tokens_ids_tensor.cuda(), attn_mask.cuda()
```

```python
def C1Classifier(Q, A):
    tokens_ids_tensor, attn_mask = preproccesing(Q, A)
    cont_reps = beto_model(tokens_ids_tensor.unsqueeze(0), attention_mask = attn_mask.unsqueeze(0))
    cls_rep = cont_reps.last_hidden_state[:, 0]
    logits = cls_layer(cls_rep)
    probs = torch.sigmoid(logits)
    return probs.detach().cpu().numpy()[0]
```

```python
def QClassifier(Q):
    tokens_ids_tensor, attn_mask = preproccesingQ(Q)
    cont_reps = beto_model(tokens_ids_tensor.unsqueeze(0), attention_mask = attn_mask.unsqueeze(0))
    cls_rep = cont_reps.last_hidden_state[:, 0]
    logits = cls_layer_Q(cls_rep)
    probs = torch.sigmoid(logits)
    return probs.detach().cpu().numpy()[0]
```
