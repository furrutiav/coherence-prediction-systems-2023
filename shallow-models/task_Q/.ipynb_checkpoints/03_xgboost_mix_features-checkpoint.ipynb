```python
import pandas as pd
import pickle
import numpy as np
```

## Data

```python
beto_pt_features_train = pd.read_excel("../../feature-engineering/features/beto_pt_features_train_task_Q.xlsx", index_col="id")
beto_pt_features_val = pd.read_excel("../../feature-engineering/features/beto_pt_features_val_task_Q.xlsx", index_col="id")
beto_pt_features_test = pd.read_excel("../../feature-engineering/features/beto_pt_features_test_task_Q.xlsx", index_col="id")
```

```python
lf_features_train = pd.read_excel("../../feature-engineering/features/lf_features_train_task_Q.xlsx", index_col="id")
lf_features_val = pd.read_excel("../../feature-engineering/features/lf_features_val_task_Q.xlsx", index_col="id")
lf_features_test = pd.read_excel("../../feature-engineering/features/lf_features_test_task_Q.xlsx", index_col="id")
```

```python
beto_pt_features_train.columns = [f"BETO<&>{c}" for c in beto_pt_features_train.columns] 
beto_pt_features_val.columns = [f"BETO<&>{c}" for c in beto_pt_features_val.columns] 
beto_pt_features_test.columns = [f"BETO<&>{c}" for c in beto_pt_features_test.columns] 
```

```python
Q_train = pd.read_excel("../../data/train_task_Q.xlsx", index_col="id")
Q_val = pd.read_excel("../../data/val_task_Q.xlsx", index_col="id")
Q_test = pd.read_excel("../../data/test_task_Q.xlsx", index_col="id")
```

```python
features_train = pd.concat([lf_features_train, beto_pt_features_train], axis=1)
features_val = pd.concat([lf_features_val, beto_pt_features_val], axis=1)
features_test = pd.concat([lf_features_test, beto_pt_features_test], axis=1)
```

```python
X_train, y_train = features_train, Q_train["label"]
X_train = X_train.loc[y_train.index]

X_test, y_test = features_test, Q_test["label"]
X_test = X_test.loc[y_test.index]

X_val, y_val = features_val, Q_val["label"]
X_val = X_val.loc[y_val.index]
```

```python
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectPercentile, f_classif
from sklearn.compose import ColumnTransformer
from sklearn.decomposition import PCA
```

```python
beto_cols = list(beto_pt_features_train.columns)
other_cols = list(set(features_train.columns).difference(set(beto_cols)))
```

```python
%%time
ct = ColumnTransformer(transformers=[
    ("selector", SelectPercentile(f_classif, percentile=50), other_cols), 
    ("pca", PCA(n_components=2), beto_cols)
])

pipe = Pipeline(steps=[
    ("ct", ct),
    ("clf", SVC(
        probability=True, 
        random_state=2022
    )),
])
```

```python
pipe.get_params()
```

```python
%%time
pipe.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
768/8
```

```python
parameters = {
    'clf__kernel': ['linear', 'rbf', "sigmoid", "poly"], 
    "clf__class_weight": ["balanced", None],
    "clf__gamma": ["scale", "auto"],
    'clf__C':[1, 10, 100, 1000],
    "ct__selector__percentile": [50, 75, 90, 100],
    "ct__pca__n_components": [768, 384, 192, 96]
}
```

```python
clf = GridSearchCV(pipe, parameters, scoring="f1_macro", cv=5, verbose=1)
```

```python
%%time
clf.fit(X_train, y_train)
```

```python
clf.best_estimator_
```

```python
y_pred = clf.predict(X_val)
print(classification_report(y_val, y_pred))
```

```python
y_pred = clf.predict(X_test)
print(classification_report(y_test, y_pred))
```
